```python
import nlpcloud



class NLPApp:
    def __init__(self):
        self.__database={} #making this attribute private
        self.__first_menu()
    
    def __first_menu(self): #privatemethod
        first_input=input(""" Hi!
        How would you like to proceed?
        1.Not a member? Register
        2.Already a member? Login
        3.Exit """)
        
        if first_input=='1':
            self.__register()
        elif first_input=='2':
            self.__login()
        else:
            pass
        
    
    
    def __register(self):
        name=input('enter name')
        email=input('Enter email')
        password=input('Enter password')
        
        if email in self.__database:
            print("Email already exists")
        else:
            self.__database[email]=[name,password]
            print("Registration successfull.Now Login")
            print(self.__database)
            self.__first_menu()
        
    
    def __login(self):
        email=input("Enter email")
        password=input("Enter password")
        
        if email in self.__database:
            if self.__database[email][1] == password:
                print('login successful')
                self.__second_menu()
            else:
                print('Wrong password.Try again')
                self.__login()
        else:
            print("This email is not registered")
            self.__first_menu()
        
        
    def __second_menu(self):
        second_input=input(""" Hi!
        How would you like to proceed?
        1.NER
        2.Language detection
        3.Sentimental analysis 
        4. Logout""")
        
        if second_input=='1':
            self.__ner()
        elif second_input=='2':
            self.__language_detection()
        elif second_input=='3':
            self.__sentiment_analysis()
        else:
            pass
        
        
    def __ner(self):
        para=input("""Enter the paragraph""")
        search_term=input("""What would you like to search""")
        client = nlpcloud.Client("finetuned-llama-3-70b", "0ccb349279e6f8d0958dbb7eed47f5e9f72ea2c0", gpu=True)
        response=client.entities(para,searched_entity=search_term)
        print(response)
        
    def __sentiment_analysis(self):
        para=input("""Enter the paragraph""")
        import nlpcloud
        client = nlpcloud.Client("distilbert-base-uncased-emotion", "0ccb349279e6f8d0958dbb7eed47f5e9f72ea2c0", gpu=False)
        response=client.sentiment(para)
        print(response)
    
    def __language_detection(self):
        para=input("""Enter the paragraph""")
        search_term=input("""What would you like to search""")
        import nlpcloud

        client = nlpcloud.Client("python-langdetect", "0ccb349279e6f8d0958dbb7eed47f5e9f72ea2c0", gpu=False)
        response=client.lang_detection(para)
        print(response)
    
    
        
            
        
        
        
        
        
        
    
        
```
